### 2.1 Замена типов данных

«Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута. Необходимо округлить данные о длительности разговора до минут в большую сторону.

Будем считать, что счет пользователю выставляется каждый месяц с момента регистрации.
У всех пользователей отличается дата подключения тарифа => у всех пользователей будет отличаться рассчетный период.
Т.е если пользователь подключился к тарифу 12.06.2018, то потраченные минуты, сообщения и мегабайты надо считать до 12 числа каждого следующего месяца.

```python
import math

# Округляем длительность разговора в большую сторону до целого
def ceil(row):
    return math.ceil(row['duration'])
df_calls['ceiled'] = df_calls.apply(ceil, axis=1)
df_calls['duration'] = df_calls['ceiled']

# Функия для добавления дня месяца
def reg_date_day(row):
    return int(row['reg_date'][8:10])
df_users['reg_date_day']=df_users.apply(reg_date_day, axis=1)

# Создаем словарь user_id:reg_date
users_date_dict = dict( zip(df_users['user_id'],df_users['reg_date_day'] ))

# Заменяем во всех таблицах имена столбцов date для более удобной обработки

df_calls = df_calls.rename(columns={"call_date": "date"})
df_messages = df_messages.rename(columns={"message_date": "date"})
df_internet = df_internet.rename(columns={"session_date": "date"})

def calc_period(row):
    user_id = row['user_id']
    date_day = int(row['date'][8:10])
    date_month = int(row['date'][5:7])
    # День подключения
    reg_day = users_date_dict.get(user_id)
    if date_day < reg_day:
        return date_month-1
    else:
        return date_month

# Применяем функцию. Получили для каждого события рассчетный месяц    
df_calls['month']=df_calls.apply(calc_period, axis=1)
df_messages['month'] = df_messages.apply(calc_period, axis=1)
df_internet['month'] = df_internet.apply(calc_period, axis=1)
```

### 2.4 Добавление данных

```python
#df_users = df_users.set_index('user_id')
# создаем таблицы с распределением по месяцам, в качестве индексов - user_id
# таблица для количества звонков
minutes_used_by_month = df_calls.query('duration != 0 and month != 13').groupby(['user_id', 'month']).sum().unstack()
# таблица для потраченных минут
calls_count_by_month = df_calls.query('duration != 0 and month != 13').groupby(['user_id', 'month'])['id'].count().unstack()
calls_count_by_month.columns = pd.MultiIndex.from_product([["count_calls"], calls_count_by_month.columns])
# таблица для количества смс
messages_count_by_month = df_messages.query('month != 13').groupby(['user_id', 'month'])['id'].count().unstack()
messages_count_by_month.columns = pd.MultiIndex.from_product([["count_messages"], messages_count_by_month.columns])
# таблица для потраченных мегабайт
mb_used_by_month = df_internet.query('month != 13').groupby(['user_id', 'month']).sum().unstack()

# создаем финальный датафрейм со всеми данными
df_total = df_users.set_index('user_id').join(calls_count_by_month, on='user_id', how='left')\
.join(minutes_used_by_month, how='left')\
.join(messages_count_by_month, how='left')\
.join(mb_used_by_month, how='left')

# функция для определения выручки по мясяцам
def revenue(row, month):
    tariff = row['tariff']
    count_minutes = row['duration', month]
    count_messages = row['count_messages', month]
    count_calls = row['count_calls', month] 
    mb_used = row['mb_used', month]
    
        
    revenue = df_tariffs.loc[tariff]['rub_monthly_fee']
    # считаем, сколько осталось минут, смс и мб
    minutes_left = df_tariffs.loc[tariff]['minutes_included'] - count_minutes
    messages_left = df_tariffs.loc[tariff]['messages_included'] - count_messages
    mb_left = df_tariffs.loc[tariff]['mb_per_month_included'] - mb_used
    # если вышел за лимит - добавляем к выручке
    if minutes_left < 0:
        revenue += abs(minutes_left)*df_tariffs.loc[tariff]['rub_per_minute']
    if messages_left < 0:   
        revenue += abs(messages_left)*df_tariffs.loc[tariff]['rub_per_message']
    if mb_left < 0:
        revenue += abs(mb_left)*df_tariffs.loc[tariff]['rub_per_gb']/1024
    return revenue
# применяем функцию
for i in range(1,13):
    df_total['revenue',i] = df_total.apply(revenue, args=(i,), axis=1)
    
df_total.head(15)
# Добавим колонки со средним значением минут, смс и трафика в месяц
# 12 месяц не используем, т.к. он еще не закончился
#def minutes_mean(row):
#    return row.iloc[17:28].mean()
#def messages_mean(row):
#    return row.iloc[29:40].mean()
#def megabytes_mean(row):
#    return row.iloc[41:52].mean()
#def revenue_mean(row):
#    return row.iloc[53:64].mean()

#df_total['min_mean'] = df_total.apply(minutes_mean, axis=1)
#df_total['msg_mean'] = df_total.apply(messages_mean, axis=1)
#df_total['mb_mean'] = df_total.apply(megabytes_mean, axis=1)
#df_total['revenue_mean'] = df_total.apply(revenue_mean, axis=1)
```

## Часть 4. Проверка гипотез.
<a id='part4.1'></a>
### 4.1 Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различается.

Сформулируем гипотезы.

H<sub>0</sub>: Средние выручки пользователей "Ультра" и "Смарт" не равны.

H<sub>1</sub>: Средние выручки пользователей "Ультра" и "Смарт" равны.


```python
from scipy import stats as st
alpha = .01

series_smart = df_total.query('tariff == "smart"')[('revenue',1)]
series_ultra = df_total.query('tariff == "ultra"')[('revenue',1)]
for i in range(2,12):
        #print(sample[columns,i])
    series_smart = series_smart.append(df_total.query('tariff == "smart"')[('revenue',i)])
    series_ultra = series_ultra.append(df_total.query('tariff == "ultra"')[('revenue',i)])

test_result = st.ttest_ind(
        series_smart, 
        series_ultra)

print(test_result)
print('p-значение: ', test_result.pvalue)

if (test_result.pvalue > alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### 4.2 Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.
Сформулируем гипотезы.

H<sub>0</sub>: Средняя выручка пользователей из Москвы равна выручке пользователей из других регионов.

H<sub>1</sub>: Средняя выручка пользователей из Москвы не равна выручке пользователей из других регионов.


```python
def hyp(tariff,alpha):
    series_moscow=[]
    series_no_moscow=[]
    series_smart = df_total.query('tariff != @tariff and city == "Москва"')[('revenue',1)]
    series_ultra = df_total.query('tariff != @tariff and city != "Москва"')[('revenue',1)]
    for i in range(2,12):
        #print(sample[columns,i])
        series_moscow = series_smart.append(df_total.query('tariff != @tariff and city == "Москва"')[('revenue',i)])
        series_not_moscow = series_ultra.append(df_total.query('tariff != @tariff and city != "Москва"')[('revenue',i)])

    test_result = st.ttest_ind(
        series_moscow, 
        series_not_moscow)


    print('p-значение: ', test_result.pvalue)
    if (test_result.pvalue < alpha):
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")

print('Тариф Ультра')
hyp('smart',0.05)
print('\nТариф Смарт')
hyp('ultra',0.05)
print('\nОба тарифа')
hyp('111',0.05)
```

Гипотезу о равенстве выручки пользователей из Москвы и регионов также не удалось опровергнуть.
